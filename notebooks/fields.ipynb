```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
%matplotlib inline

from astropy.table import Table
import astropy.units as u
import astropy.coordinates as coord
from astropy.io import ascii

from pyia import GaiaData
import gala.coordinates as gc
```

```python
plt.style.use('notebook')
```

```python
g = GaiaData('../data/gd1-with-masks.fits')
```

```python
stream = g[g.pm_mask & g.gi_cmd_mask & (g.phi2>-5*u.deg)]
```

```python
fields_phi1 = np.array([-32.77, -32.77, -46.5])
fields_phi2 = np.array([1.167, 0, -0.1])
```

```python
plt.figure(figsize=(12,7))
plt.plot(stream.phi1, stream.phi2, 'ko', ms=2.5, alpha=0.7, mec='none')
plt.plot(fields_phi1, fields_phi2, 'o', color='tab:orange', mec='k', mew=1, ms=11, alpha=0.6,
         label='Proposed fields')

ctext = '0.2'

plt.annotate('Spur', xy=(-33, 2), xytext=(-26, 7), ha='center',
             arrowprops=dict(color='0.3', shrink=0.08, width=1.5, headwidth=6, headlength=8, alpha=0.4),
             fontsize='medium')
plt.text(-23, 7, 'of ejected GD-1 stars?', fontsize='small', color=ctext, ha='left')

plt.annotate('Gap', xy=(-41, -1.5), xytext=(-41, -8), ha='center',
             arrowprops=dict(color='0.3', shrink=0.08, width=1.5, headwidth=6, headlength=8, alpha=0.4),
             fontsize=20)
plt.text(-38, -8, 'induced by a dark matter subhalo?', fontsize='small', color=ctext, ha='left')

plt.annotate('Overdensity', xy=(-48, 1), xytext=(-52, 9), ha='right',
             arrowprops=dict(color='0.3', shrink=0.08, width=1.5, headwidth=6, headlength=8, alpha=0.4),
             fontsize=20)
plt.text(-52, 6.5, 'a stream fold?', fontsize='small', color=ctext, ha='right')

plt.gca().set_aspect('equal')
plt.xlim(-80,0)
plt.ylim(-12,12)

plt.legend(loc=3, frameon=False, fontsize='small', handlelength=0.7)
plt.xlabel('$\phi_1$ [deg]')
plt.ylabel('$\phi_2$ [deg]')

plt.tight_layout()
plt.savefig('../plots/fields.pdf')
```
